```python
import pandas as pd
import seaborn as sns

df = pd.read_excel("Man-Whitney W2-W0.xlsx")
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mannwhitneyu

# Definir las columnas de interés y los grupos correspondientes
columns_of_interest = ['ΔVolumen S2', 'DSC S2', 'PPV S2', 'Sens S2', 'PC S2']
groups = [
    ('Recurrencia Local', df['Recurrencia Local']),
    (' Metastasis a Distancia', df[' Metastasis a Distancia']),
    ('Progresion', df['Progresion']),
    ('Promedio de Supervivencia ', df['Promedio de Supervivencia '])
]

# Función para calcular y mostrar los resultados de Mann-Whitney U y generar el gráfico de cajas si p < 0.05
def calculate_mannwhitneyu_and_boxplot(df, col, group1, group2, group_name):
    stat, p_value = mannwhitneyu(df[group1][col], df[group2][col])
    print(f'Estadístico U {col}: {stat}')
    print(f'Valor p {col}: {p_value}')
    
    if p_value < 0.05:
        plt.figure(figsize=(8, 6))
        sns.boxplot(x=group_name, y=col, data=df, palette='Set2')
        plt.title(f'Comparación entre {col} y {group_name}')
        plt.xlabel(group_name)
        plt.ylabel(col)
        plt.show()

# Iterar sobre los grupos y columnas de interés
for group_name, group_data in groups:
    print(f"**RESULTS {group_name.upper()}**")  # Línea de separación
    for col in columns_of_interest:
        calculate_mannwhitneyu_and_boxplot(df, col, group_data == 1, group_data == 0, group_name)
    print("-" * 50)  # Línea de separación
```
