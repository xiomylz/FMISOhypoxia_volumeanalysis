```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

data = pd.read_excel("Correlaciones W2-W0.xlsx")
data
```

```python
data = data.drop('FMISO', axis=1)
correlation= data.corr()
sns.heatmap(correlation, annot= True)
```

```python
from itertools import combinations
# Crear una lista para almacenar los resultados
results_list = []

# Iterar sobre todas las combinaciones de columnas sin repetir y sin auto-correlaciones
for col1 in data.columns:
    for col2 in data.columns:
        r, p = pearsonr(data[col1], data[col2])
        # Crear un DataFrame temporal para cada resultado
        temp_df = pd.DataFrame({'Column1': [col1], 'Column2': [col2], 'r': [r], 'p': [p]})
        # AÃ±adir el DataFrame temporal a la lista
        results_list.append(temp_df)

# Concatenar todos los DataFrames de la lista en uno solo
results_df = pd.concat(results_list, ignore_index=True)

print(results_df.head(500))
```

```python
# Calcular los valores p para todas las combinaciones de columnas
corr_p_values = pd.DataFrame(index=data.columns, columns=data.columns)

for col1 in data.columns:
    for col2 in data.columns:
        r, p = pearsonr(data[col1], data[col2])
        corr_p_values.loc[col1, col2] = p

# Crear un heatmap utilizando seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(corr_p_values.astype(float), annot=True, cmap='coolwarm', vmin=0, vmax=0.05, fmt='.3f', linewidths=.5)
plt.title('Heatmap de Valores p de CorrelaciÃ³n')
plt.show()
```

```python
correlation.to_excel('correlations_table_w2.xlsx', sheet_name='Correlations W2')
```

```python
correlacion_editada = correlation.applymap(lambda x: 'SÃ­' if x > 0.7 else 'No')
correlacion_editada
```

```python
correlacion_editada.to_excel('correlations_table_w2_yes_no.xlsx', sheet_name='Correlations W2')
```
