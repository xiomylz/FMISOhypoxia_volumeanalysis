```python
import pandas as pd
from lifelines.statistics import logrank_test
from lifelines import KaplanMeierFitter
import matplotlib.pyplot as plt

# Leer los datos del archivo Excel
file_path = r'C:\Users\xiomy\OneDrive\Escritorio\TFM FISICA MEDICA\medical-image-analysis\Test Long-Rank y Curvas Kaplan Meier\Volumen S2.xlsx'
data = pd.read_excel(file_path)

# Calcular la mediana de los valores en la columna 'CP W2'
cp_values = data['ΔVolumen S2']
median_cp = cp_values.median()

# Clasificar los datos en dos grupos: menores y mayores a la mediana
group_low = data[data['ΔVolumen S2'] <= median_cp]
group_high = data[data['ΔVolumen S2'] > median_cp]

# Calcular los meses promedio de progresión para cada grupo
avg_months_low = group_low['Months  P'].mean()
avg_months_high = group_high['Months  P'].mean()

# Realizar el test Log-Rank
results = logrank_test(group_low['Months  P'], group_high['Months  P'], event_observed_A=None, event_observed_B=None)

# Mostrar los resultados
print("Mediana de CP W2:", median_cp)
print("Resultados del test Log-Rank:")
print(results)
print(results.summary)
print("Meses promedio de progresión para CP <= Mediana:", avg_months_low)
print("Meses promedio de progresión para CP > Mediana:", avg_months_high)

# Inicializar el objeto KaplanMeierFitter
kmf_low = KaplanMeierFitter()
kmf_high = KaplanMeierFitter()

# Ajustar el modelo Kaplan-Meier para ambos grupos
kmf_low.fit(group_low['Months  P'], label=f'ΔVolumen <= Mediana (Promedio: {avg_months_low:.2f} meses)')
kmf_high.fit(group_high['Months  P'], label=f'ΔVolumen > Mediana (Promedio: {avg_months_high:.2f} meses)')

# Graficar las curvas de Kaplan-Meier
plt.figure(figsize=(10, 6))
kmf_low.plot_survival_function()
kmf_high.plot_survival_function()

# Añadir detalles al gráfico
plt.title('Curvas de Kaplan-Meier para ΔVolumen en S2')
plt.xlabel('Tiempo en meses')
plt.ylabel('Probabilidad de supervivencia de progresión')
plt.text(0.6, 0.6, f'Log-Rank Test p-value: {results.p_value:.4f}', transform=plt.gca().transAxes)

# Mostrar el gráfico
plt.show()
```
